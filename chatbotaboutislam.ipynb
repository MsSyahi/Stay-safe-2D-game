```python
#Import the libraries
from newspaper import Article
import random
import string
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#Download the punkt package
nltk.download('punkt', quiet=True)
```

```python
#Get article
article = Article('https://en.wikipedia.org/wiki/Islam')
article.download()
article.parse()
article.nlp()
total_text = article.text
```

```python
#Print the articles text
print(total_text)
```

```python
text = total_text
sentences_list = nltk.sent_tokenize(text)
print(sentences_list)
```

```python
from nltk import tokenize
#Tokenization
text = total_text
sentences_list = nltk.sent_tokenize(text) # A list of sentences
```

```python
#Print the list of sentences
print(sentences_list)
```

```python
#a function to return a random greeting response to a users greeting
def greeting_response(text):
  text = text.lower()

  #bots greeting response
  bot_greetings = ['waalaikumussalam', 'hai', 'hello', 'hola']

  #user greeting 
  user_greetings = ['assalamualaikum', 'salam', 'hai', 'hello', 'hola']

  for word in text.split():
    if word in user_greetings:
      return random.choice(user_greetings)
```

```python
#for sorting list

def index_sort(list_var):
  length = len(list_var)
  list_index = list(range(0, length))

  x = list_var
  for i in range(length):
    for j in range(length):
      if x[list_index[i]] > x[list_index[j]]:
        #swap
        temp = list_index[i]
        list_index[i] = list_index[j]
        list_index[j] = temp

  return list_index
   
```

```python
#create the bots response
def bot_response(user_input):
  user_input = user_input.lower()
  sentences_list.append(user_input)
  bot_reponse = ''
  CountM = CountVectorizer().fit_transform(sentences_list)
  similarity_scores = cosine_similarity(CountM[-1], CountM)
  similarity_scores_list = similarity_scores.flatten()
  index = index_sort(similarity_scores_list)  
  index = index[1:]
  response_flag = 0

  j = 0
  for i in range(len(index)):
    if similarity_scores_list[index[i]] > 0.0:
      bot_response = bot_response + ' ' + sentences_list[index[i]]
      response_flag = 1
      j = j + 1
    if j > 3:
      break

    if response_flag == 0:
      bot_response = bot_response + ' ' + 'Sorry I dont understand.'

    sentences_list.remove(user_input)

    return bot_response
```

```python
#start the chat
print('Syahi Bot: Assalamualaikum I am Syahi, I will answer your question regarding Islam. Let me know what you want to know about Islam')

exit_list = ['exit', 'see you later', 'i go first', 'thank you']
while(True):
  user_input = input()
  if user_input.lower() in exit_list:
    print('Syahi Bot: Byee, have a nice day ! ')
    break
  else:
    if greeting_response(user_input) != None:
      print('Syahi Bot: ' + greeting_response(user_input)) 
    else:
      print('Syahi Bot: ' + bot_response(user_input))
      

```
